```python
X_BMI = mydata.iloc[:, 0:3]
```

## Binary Encoding

```python
X_BMI['Gender'] = X_BMI['Gender'].map({'Male':0, 'Female':1})
```

```python
Y_TARGET = mydata.iloc[:, 3]
```

```python
X_BMI.values
```

```python
X_BMI.head()
```

```python
X_INPUT = X_BMI.values
Y_INPUT = Y_TARGET.values
```

```python
X_TRAIN = X_INPUT[:400]
X_TEST = X_INPUT[400:]
```

```python
Y_TRAIN = Y_INPUT[:400]
Y_TEST = Y_INPUT[400:]
```

```python
'samples in {0} : {1}'.format('X_TRAIN', len(X_TRAIN)), 'samples in {0} : {1}'.format('X_TEST', len(X_TEST))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(X_TRAIN, Y_TRAIN)
```

```python
result = model.predict([[0, 160, 40], [1, 170, 70]])
```

```python
index_name[result]
```

```python
model.score(X_TEST, Y_TEST)
```

```python
YALIST = list(Y)
YPLIST = model.predict(X_TRAIN)
```

```python
difres = pd.DataFrame({'YA':YALIST, 'YP':YPLIST})
difres.head()
```

```python
difres[difres['YA'] != difres['YP']].head()
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(YALIST, YPLIST)
acc
```

```python
def model_test(neighbors = 5):
    model = KNeighborsClassifier(n_neighbors=neighbors)
    model.fit(X_TRAIN, Y_TRAIN)
    return model.score(X_TEST, Y_TEST)
```

```python
for idx, val in enumerate([model_test(i) for i in range(1, 10)]):
    print('Neighbors : {0}, Accuracy : {1}'.format(idx+1, val))
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 400, 400)
y = [model_test(i) for i in range(1, 401)]
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.title('Curve')
plt.text(10, 0.96, '  (10, 0.96)')
plt.grid()
plt.plot([10], [0.96], marker='o')
plt.show()
```
